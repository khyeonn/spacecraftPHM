```python
from minisom import MiniSom
import numpy as np
import matplotlib.pyplot as plt

# Define the SOM grid dimensions (you can adjust these as needed)
grid_size = (20, 20)  # You can change this based on your preference

# Create and initialize the SOM
som = MiniSom(grid_size[0], grid_size[1], X_train.shape[1], sigma=1.0, learning_rate=0.5, random_seed=42)

# Train the SOM using only healthy data
som.train(X_train, 1000, verbose=True)

# Calculate the U-matrix manually for the test data
u_matrix_test = np.zeros((grid_size[0], grid_size[1]))
mqe_list = []
for i, test_point in enumerate(X_test):
    bmu = som.winner(test_point)  # Find the best-matching unit for the test point
    # Calculate the distances to the neighboring neurons
    array = []
    for x in range(grid_size[0]):
        for y in range(grid_size[1]):
            weight = som.get_weights()[x, y]
            distance = np.linalg.norm(weight - test_point)
            u_matrix_test[x, y] += distance
            array.append(distance)
    mqe = np.mean(array)
    mqe_list.append(mqe)
# Normalize the U-matrix for the test data
u_matrix_test /= u_matrix_test.max()

# Create a U-matrix plot (HIT map) for the test data
plt.figure(figsize=(10, 8))
plt.pcolor(u_matrix_test.T, cmap='coolwarm')
plt.colorbar()

# Title and labels for the plot
plt.title('U-Matrix for Test Data')
plt.xlabel('X-Axis Neuron Index')
plt.ylabel('Y-Axis Neuron Index')

plt.show()




max_mqe = max(mqe_list)
mqe_list = [mqe / max_mqe for mqe in mqe_list]
print(mqe_list)
# Create a plot of MQE for each test trial
plt.figure(figsize=(10, 4))
plt.plot(range(len(X_test)), mqe_list, marker='o', linestyle='-')
plt.title('Mean Quantization Error (MQE) for Test Data')
plt.xlabel('Sample Number')
plt.ylabel('MQE')
plt.grid(True)
plt.show()
```
